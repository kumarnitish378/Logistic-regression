```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = -4
X = []
Y = []
while x*x <= 16:
    X.append(x)
    Y.append(x*x)
    x = x+0.001
plt.plot(X,Y, color="red")
plt.grid()
plt.show()
```

```python
x = -4
X = []
Y = []
while (x*x + 2*(x*x*x)) <= 112:
    X.append(x)
    Y.append(x*x + 2*(x*x*x))
    x = x+0.001
plt.plot(X,Y, color="red")
plt.grid()
plt.show()
```

```python
# circle r^2 = sqrt(y^2 +x^2)
import math
x = 5
r = 5
Y = []
X = []
while x >= -5 :
    y = math.sqrt((r*r) - (x*x))
    X.append(x)
    Y.append(y)
    x = x-0.001
# print(X)
# print(Y)
plt.plot(X,Y, color="red")
plt.grid()
plt.show()
```

```python
# exponential graph 'y = 2x'
x = 2
Y = []
X = []
while x <= 10 :
    y = math.pow(2,x)
    X.append(x)
    Y.append(y)
    x = x+0.0001
# print(X)
# print(Y)
plt.plot(X,Y, color="red")
plt.grid()
plt.show()
```

```python
# velocity [ v= 0.5*a*t^2 ]
x = 9.8
t = 0
Y = []
X = []
while t <= 10 :
    v = 0.5*x*t*t
    X.append(t)
    Y.append(v)
    t = t+0.1
# print(X)
# print(Y)
plt.plot(X,Y, color="red")
plt.grid()
plt.show()
```

```python
time        = np.arange(0, 10, 0.1)
# print(time)
amplitude   = np.sin(20/time)
plt.plot(time, amplitude)
plt.ylabel('Amplitude = 20/sin(time)')
plt.xlabel('Time')
plt.axhline(y=amplitude.min(), color='red')
plt.grid(True, which='both')
plt.show()
```
